```python
import pandas as pd
from collections import Counter
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/Users/macbookpro/Project_construction/BIM_0514.csv')
```

```python
pubyear = list(data["pubyear"])
set_pubyear =list(set(set(pubyear)))
cnt_data = Counter(pubyear)
values = list(cnt_data.values())
    
```

```python
plt.figure(figsize= (10,6))
plt.plot(set_pubyear, values, marker = '*')
plt.yscale('log')
plt.xticks(set_pubyear, rotation = 45)
plt.xlabel('Year')
plt.ylabel('Number of articles(log scale)')
plt.title('BIM articles publications')
plt.tight_layout()
plt.savefig('/Users/macbookpro/Project_construction/BIM_log_0514')

plt.figure(figsize= (10,6))
plt.plot(set_pubyear, values, marker = '*')
plt.xticks(set_pubyear, rotation = 45)
plt.xlabel('Year')
plt.ylabel('Number of articles')
plt.title('BIM articles publications')
plt.tight_layout()
plt.savefig('/Users/macbookpro/Project_construction/BIM_naive_0514')
```
